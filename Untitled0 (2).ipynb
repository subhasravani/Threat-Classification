## Data cleaning

### Subtask:
Clean the dataset stored in the DataFrame `df`.


**Reasoning**:
Identify and handle missing values and remove duplicate rows in the dataframe.



```python
# Identify missing values
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)

# Handle missing values (replace with mean for numerical columns)
for column in df.columns:
  if df[column].dtype in ['int64', 'float64']:
    df[column].fillna(df[column].mean(), inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

print("Number of rows after removing duplicates:", df.shape[0])
```

## Data wrangling

### Subtask:
Prepare the data for model training by converting categorical features into numerical representations and performing feature scaling.


```python
import numpy as np

# Replace infinite values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop rows with NaN values
df.dropna(inplace=True)

# Identify categorical features
categorical_features = ['Label']

# Create a ColumnTransformer to apply different preprocessing steps to different columns
ct = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(handle_unknown='ignore'), categorical_features),
        ('scaler', StandardScaler(), [col for col in df.columns if col not in categorical_features])
    ],
    remainder='passthrough'
)

# Fit and transform the data
df_transformed = ct.fit_transform(df)

# Create a new DataFrame with the transformed data
df_transformed = pd.DataFrame(df_transformed)
display(df_transformed.head())
```

## Feature engineering

### Subtask:
Perform feature selection and dimensionality reduction using PCA on the transformed data stored in `df_transformed`.


**Reasoning**:
Perform feature selection and dimensionality reduction using PCA on the transformed data stored in `df_transformed`.



```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming 'Label' is the target variable
y = df['Label']

# Select features (all except 'Label')
X = df_transformed.drop(0, axis=1)

# Standardize the features
X = StandardScaler().fit_transform(X)

# Apply PCA
pca = PCA(n_components=0.95)  # Retain components explaining 95% of variance
principal_components = pca.fit_transform(X)

# Create a new DataFrame with the PCA-transformed data
df_pca = pd.DataFrame(data=principal_components)

display(df_pca.head())
```

## Data splitting

### Subtask:
Split the PCA-transformed data (`df_pca`) into training and testing sets.


**Reasoning**:
Split the PCA-transformed data into training and testing sets using train_test_split.



```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df_pca, df['Label'], test_size=0.25, random_state=42)
```

## Model training

### Subtask:
Train a Random Forest and an SVM model on the training data (`X_train`, `y_train`)


**Reasoning**:
Train the Random Forest and SVM models using the provided training data.



```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Instantiate a Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)

# Instantiate an SVM model
svm_model = SVC(kernel='linear', C=1.0, random_state=42)

# Train the Random Forest model
rf_model.fit(X_train, y_train)

# Train the SVM model
svm_model.fit(X_train, y_train)
```

## Model evaluation

### Subtask:
Evaluate the performance of the trained Random Forest and SVM models using the test data.


**Reasoning**:
Evaluate the performance of the trained Random Forest and SVM models using the test data by generating predictions and calculating the evaluation metrics.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

# Make predictions on the test data using the Random Forest model
y_pred_rf = rf_model.predict(X_test)

# Evaluate the Random Forest model
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf, average='weighted')
recall_rf = recall_score(y_test, y_pred_rf, average='weighted')
f1_rf = f1_score(y_test, y_pred_rf, average='weighted')
confusion_matrix_rf = confusion_matrix(y_test, y_pred_rf)

print("Random Forest Model Evaluation:")
print("Accuracy:", accuracy_rf)
print("Precision:", precision_rf)
print("Recall:", recall_rf)
print("F1-Score:", f1_rf)
print("Confusion Matrix:\n", confusion_matrix_rf)
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))


# Make predictions on the test data using the SVM model
y_pred_svm = svm_model.predict(X_test)

# Evaluate the SVM model
accuracy_svm = accuracy_score(y_test, y_pred_svm)
precision_svm = precision_score(y_test, y_pred_svm, average='weighted')
recall_svm = recall_score(y_test, y_pred_svm, average='weighted')
f1_svm = f1_score(y_test, y_pred_svm, average='weighted')
confusion_matrix_svm = confusion_matrix(y_test, y_pred_svm)

print("\nSVM Model Evaluation:")
print("Accuracy:", accuracy_svm)
print("Precision:", precision_svm)
print("Recall:", recall_svm)
print("F1-Score:", f1_svm)
print("Confusion Matrix:\n", confusion_matrix_svm)
print("\nClassification Report:\n", classification_report(y_test, y_pred_svm))
```
